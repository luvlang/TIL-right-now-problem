## 문제해결 프로세스 및 체크리스트
### 머신러닝 문제해결 프로세스
- 경진대회 이해
	- 문제 이해: 미션, 배경/목적, 어떤 데이터를 활용해서 어떤 값을 예측, 회귀vs분류
	- 평가 지표 파악
- 탐색적 데이터 분석
	- 데이터 구조 탐색: 구조훑기(파일별 용도, 데이터-레코드,피처-의 수량 및 이해, 훈련데이터vs테스트데이터, 타깃값 확인), 몇가지 통계
	- 데이터 시각화: 다양한 그래프 활용(수치형vs범주형, 관계 시각화) -> (인사이트) 어떤 피처가 중요한지, 피처끼리 어떻게 조합해서 새로운 피처를 만들지, 어떤 점을 주의해서 모델링할지
	- 피처/이상치 파악: 추가/제거/인코딩 전략
- 베이스라인 모델
  - 모델 훈련 및 성능 검증: 기본적인 머신러닝 파이프라인 구축, 피처 엔지니어링, 훈련-훈련 데이터 활용
	- 결과 예측 및 제출: 훈련된 모델을 활용해 결과 예측-테스트 데이터, 제출
- 성능 개선
  - 피처 엔지니어링: 이상치 제거, 결측값 처리, 데이터 인코딩, 피처 스케일링, 파생 피처(변수)생성, 피처 제거 및 선택 등을 수행
	- 하이퍼파라미터 최적화(튜닝): 모델의 최적 하이퍼파라미터를 찾는 작업 (ex.그리드서치, 랜덤서치, 베이지안 최적화)
	- 성능 검증: 훈련된 모델의 일반화 성능을 평가, 교차 검증(훈련데이터를 그룹화하여 검증)
	- 결과 예측 및 제출: 캐글에서는 최종 버전 2개를 선택해 제출

# 주요 시각화 그래프
## 데이터 종류
### 수치형 데이터 (사칙연산 가능)
- 연속형 데이터
- 이산형 데이터
### 범주형 데이터 (범주로 나누어짐)
- 순서형 데이터(rank)
- 명목형 데이터(nominal)
## 수치형 데이터 시각화
- 일정한 범위 내에서 어떻게 분포되어 있는지가 중요
- seaborn이 제공하는 주요 분포도 함수
	- histplot(): 히스토그램
	- kdeplot(): 커널밀도추정 함수 그래프
	- displot(): 붐포도
	- rugplot(): 러그플롯

```python
import seaborn as sns

titanic = sns.load_dataset('titanic')
```

```python
titanic.head()
```

### 히스토그램 (histplot)
- 수치형 데이터의 구간별 빈도수를 나타내는 그래프

```python
sns.histplot(data=titanic, x='age')
```

```python
#bins 파라미터의 기본값=auto(20개 구간)
sns.histplot(data=titanic, x='age', bins=10)
```

```python
#범주형 데이터의 분포가 서로 겹치도록 그려짐
sns.histplot(data=titanic, x='age', hue='alive')
```

```python
#범주형 데이터의 분포가 누적해서 그려짐
sns.histplot(data=titanic, x='age', hue='alive', multiple='stack')
```

### 커널 밀도 추정 함수 그래프 (kdeplot)
- 히스토그램을 매끄럽게(연속적으로) 곡선으로 연결한 그래프

```python
sns.kdeplot(data=titanic, x='age')
```

```python
# hue, multiple 파라미터
sns.kdeplot(data=titanic, x='age', hue='alive', multiple='stack');
```

### 분포도 (displot)
- 수치형 데이터 하나의 분포를 나타내는 그래프
- 파라미터 조정시 hisplot(), kdeplot()이 제공하는 기본 그래프를 그릴 수 있음.
