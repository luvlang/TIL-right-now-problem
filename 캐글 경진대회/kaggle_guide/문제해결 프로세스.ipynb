## 데이터 경진대회 문제를 해결하는 절차
- 대회 문제 이해 -> 데이터 분석(탐색) -> 베이스라인 모델 -> 성능 개선
### 머신러닝 경진대회
1. 대회 문제/평가지표 이해
	- 미션, 배경/목적, 어떤 데이터를 활용해서 어떤 값을 예측, 회귀vs분류
2. 데이터 분석(탐색)
	- 구조 훑기/통계: 파일별 용도, 데이터-레코드,피처-의 수량 및 이해, 훈련데이터 vs 테스트데이터, 타깃값 확인
	- 데이터 시각화: 다양한 그래프 활용(수치형vs범주형, 관계 시각화) -> (인사이트) 어떤 피처가 중요한지, 피처끼리 어떻게 조합해서 새로운 피처를 만들지, 어떤 점을 주의해서 모델링할지
	- 피처/이상치 파악: 추가/제거/인코딩 전략
3. 베이스라인 모델  
	- 모델 훈련 및 성능 검증: 기본적인 머신러닝 파이프라인 구축(데이터 불러오기), 피처 엔지니어링, 평가지표 계산함수, 훈련(훈련 데이터), 성능 검증(검증 데이터)
	- 결과 예측 및 제출: 훈련된 모델을 활용해 결과 예측-테스트 데이터, 제출
4. 성능 개선
	- 피처 엔지니어링: 이상치 제거, 결측값 처리, 데이터 인코딩, 피처 스케일링, 파생 피처(변수)생성, 피처 제거 및 선택, 데이터 다운 캐스팅, 데이터 조합 생성 등을 수행
	- 하이퍼파라미터 최적화(튜닝): 하이퍼파라미터 파악, 모델의 최적/고정 하이퍼파라미터를 찾는 작업, 값 범위 설정, 기법 선택 및 훈련(ex.그리드서치, 랜덤서치, OOF 예측, 베이지안 최적화)
	- 성능 검증: 훈련된 모델의 일반화 성능을 평가, 교차 검증(훈련데이터를 그룹화하여 검증)
	- 결과 예측 및 제출: 캐글에서는 최종 버전 2개를 선택해 제출

### 딥러닝 경진대회
1. 대회 문제/평가지표 이해
	- 문제 이해: 배경/목적, 어떤 데이터를 활용해서 어떤 값을 예측, 이진vs다중 분류/기타
2. 데이터 분석(탐색)
	- 데이터 구조 탐색: 주로 이미지, 약간의 메타데이터
		- 파일별 용도, 데이터의 양, 피처 이해, 훈련데이터vs테스트데이터, 타깃값 이해
	- 데이터 시각화
    - 타겟값 분포(시각화/이미지 데이터 출력->불량 이미지 포함 여부 체크)
    - 충분한 데이터 수 확보
3. 베이스라인 모델
	- 환경 설정: 시드값 고정(재현성O), GPU 설정
	- 데이터 준비(사용할 딥러닝 프레임워크에 맞춤): 훈련/검증 데이터 분리, 데이터셋 클래스 정의, 데이터 증강(변환기-torchvision,Albumentations 등), 데이터셋 생성, 데이터 로더
	- 모델 훈련 및 성능 검증: 기본적인 딥러닝 파이프라인 구축(신경망 모델을 직접 설계 vs 사전 훈련 모델을 불러와 사용), 훈련 전에 손실함수와 옵티마이저를 설정, 훈련 및 예측(검증 데이터), 평가
	- 결과 예측 및 제출: 훈련된 모델을 활용해 결과 예측(테스트 데이터), 제출
4. 성능 개선
	- 성능 개선 기법 적용: (이미지 변환기 활용한) 데이터 증강, 모델 선정 및 개선, 훈련 단계-옵티마이저/스케줄러/손실함수/에폭 수 재설정, 테스트 단계-데이터 증강/레이블 스무딩, 하이퍼파라미터 조정, 모델 간의 앙상블
	- 성능 검증: 훈련된 모델의 일반화 성능을 평가(검증 데이터로 예측), 성능이 좋지 않다면 이해/분석/검증 기법 단계로 되돌아가 반복
	- 결과 예측 및 제출: 캐글에서는 최종 버전(테스트 데이터로 예측) 2개를 선택해 제출

## 정형 데이터 종류/시각화
- matplotlib/seaborn 라이브러리 활용

```python
# seaborn의 load_dataset(데이터셋명) 함수 활용 -> dataframe을 반환
import seaborn as sns

titanic = sns.load_dataset('titanic')
```

- 사칙 연산이 가능한 **수치형 데이터**
	- 종류
		- 연속형 데이터: 값이 연속된 데이터
		- 이산형 데이터: 정수와 같이 값이 이산적인 데이터
	- 분포도 함수

```python
# 1. 히스토그램(histplot)
# 수치형 데이터의 구간별 빈도수를 나타내는 그래프
# 파라미터 -> data=데이터프레임, x=파악하고자 하는 피처, bins=구간 개수, hue=범주형 피처, mutiple=누적 표현인 경우 'stack'
sns.histplot(data=titanic, x='age')
```

```python
# 2. 커널 밀도 추정 함수(kedplot)
# 커널함수를 이용한 밀도추정 방법
# - 밀도 추정: 어떤 변수가 가질 수 있는 값 및 그 값을 가질 가능성의 정도를 추정하는 것. 어떤 변수의 확률밀도 함수를 추정하는 것과 동일한 말.
# - 커널 함수: 원점을 중심으로 좌우 대칭, 적분값=1, non-negative 함수
# - 각 데이터를 커널함수로 대치하여 더함으로써 non-parametric 밀도추정의 가장 단순한 예인 히스토그램의 불연속성을 해결.
# 파라미터 -> data=데이터프레임, x=파악하고자 하는 피처, hue=범주형 피처, multiple=누적 여부 등('stack')
sns.kdeplot(data=titanic, x='age')
```

```python
# 3. 분포도(displot)
# 수치형 데이터 하나의 분포를 나타내는 그래프. 파라미터에 따라 기본 그래프를 그릴 수 있음.
# 파라미터 -> data=데이터프레임, x=피처
# kind 파라미터 -> 기본값=히스토그램, 'kde'=커널밀도추정
# kde 파라미터 -> 불리언 -> 동시에 그릴 수 있음
sns.displot(data=titanic, x='age', kde=True)
```

```python
# 4. 러그플롯(rugplot)
# 주변 분포를 나타내는 그래프, 다른 분포도 그래프와 함께 사용
sns.kdeplot(data=titanic, x='age')
sns.rugplot(data=titanic, x='age')
```
